```python
# Kaggle
# Station 1: CHICAGO O'HARE INTERNATIONAL AIRPORT Lat: 41.995 Lon: -87.933 Elev: 662 ft. above sea level
# Station 2: CHICAGO MIDWAY INTL ARPT Lat: 41.786 Lon: -87.752 Elev: 612 ft. above sea level


training_days = train_df.Date.unique()
print len(train_df.Date.unique())

print len(weather_df.Date.unique())

weather_sub = weather_df[weather_df['Date'].isin(training_days)]
print len(weather_df), len(weather_sub)

LAT1 = 41.995
LON1 = -87.933
LAT2 = 41.786
LON2 = -87.752
```

```python
from geopy.distance import vincenty

def makeAirportPct(x):
    xcoord = x['Latitude']
    ycoord = x['Longitude']    
    L1 = np.sqrt(abs(LAT1-xcoord)**2 + abs(LON1-ycoord)**2) 
    L2 = np.sqrt(abs(LAT2-xcoord)**2 + abs(LON2-ycoord)**2)
    x['STATION1DIST'] = L1
    x['STATION2DIST'] = L2
    x['STATION1PCT'] = 0.5
    x['STATION2PCT'] = 0.5
    return x

train_df = train_df.apply(makeAirportPct,axis=1)
```
